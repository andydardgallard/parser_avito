```python
import requests
from bs4 import BeautifulSoup
from random import choice
from time import sleep
from random import  uniform
from selenium import webdriver
from PIL import Image
from pytesseract import image_to_string
import codecs
import csv

def write_csv(new_line):
    with open("town_moskovskaya_oblast_1.csv", "a" ) as fout:
        writer=csv.writer(fout)
        writer.writerow((new_line['price'],
                        new_line['price_m2'],
                        new_line['selname'],
                        new_line['phone'],
                        new_line['address'],
                        new_line['area'],
                        new_line['floors'],
                        new_line['land'],
                        new_line['centr'],
                        new_line['lat'],
                        new_line['lon'],
                        new_line['descr'],
                        new_line['url']))       
```

```python
url_spy='http://sitespy.ru/my-ip'
url='https://www.avito.ru/moskovskaya_oblast/doma_dachi_kottedzhi/prodam/taunhaus?p=1&i=1'
urls=[]
ads_list=[]

base_url='https://www.avito.ru/moskovskaya_oblast/doma_dachi_kottedzhi/prodam/taunhaus?'
page_part='p='
qyery_part='&i=1'
    
#useragents=open('useragents.txt').read().split('\n')
#proxies=open('proxy.txt').read().split('\n')
#check=1

#while check>0:
#    proxy={'http': 'http://' + choice(proxies)}
#    useragent={'User-Agent': choice(useragents)}

#    try:
#        html_spy=requests.get(url_spy, headers=useragent, proxies=proxy).text
#        soup_spy=BeautifulSoup(html_spy, 'lxml')
#        ip=soup_spy.find('span', class_='ip').text.strip()
#        ua=soup_spy.find('span', class_='ip').find_next_sibling('span').text.strip()
#        print ua
#        print useragent
#        print ip
#        print proxy  
#        check=0
#    except:
#        check=1    
    
#   if check==1:
#        print 'подключаюсь'
            
#if check==0:
html=requests.get(url).text    
soup=BeautifulSoup(html, 'lxml')
pages=soup.find('div', class_="pagination-pages").find_all('a', class_="pagination-page")[-1].get('href')
total_pages=int(pages.split('=')[1].split('&')[0])
print 'количество страниц:', total_pages   

i=1 
#while i<2:
while i<total_pages+1:
    sleep(uniform(3,5))
    url_gen=base_url+page_part+str(i)+qyery_part    
    i=i+1
    
    #try:
        #proxy={'http': 'http://' + choice(proxies)}
        #useragent={'User-Agent': choice(useragents)}
        
    #   html_spy=requests.get(url_spy).text
    #    soup_spy=BeautifulSoup(html_spy, 'lxml')
    #    ip=soup_spy.find('span', class_='ip').text.strip()
    #    ua=soup_spy.find('span', class_='ip').find_next_sibling('span').text.strip()
    #    print ua
    #    print useragent
    #    print ip
    #    print proxy 
    check=0
        
    #except:
    #    check=1
    #   i=i-1
        
    #if check==1:
    #    print 'подключаюсь'
    
    if check==0:
        html=requests.get(url_gen).text
        soup=BeautifulSoup(html, 'lxml')    
        ads=soup.find('div', class_='catalog-list').find_all('div', class_='item_table')
        print url_gen
        
        for ad in ads:            
#url_ads
            try:
                url_ads='https://www.avito.ru'+ad.find('div', class_='description').find('h3').find('a').get('href')
            except:
                url_ads=''
            urls.append(url_ads)
            print url_ads
print "количество объявлений", len(urls)
```

```python
for i in range(len(urls)):
#for i in range(5,6):
    url_ads=urls[i]   
    print url_ads
    check=1
    
    sleep(uniform(3,6))

    while check>0:
        #try:
        #    proxy={'http': 'http://' + choice(proxies)}
        #    useragent={'User-Agent': choice(useragents)}
        
        #    html_spy=requests.get(url_spy, headers=useragent, proxies=proxy).text
        #    soup_spy=BeautifulSoup(html_spy, 'lxml')
        #    ip=soup_spy.find('span', class_='ip').text.strip()
        #    ua=soup_spy.find('span', class_='ip').find_next_sibling('span').text.strip()
        #    print ua
        #    print useragent
        #    print ip
        #    print proxy   
        check=0
        #except:
        #    check=1
    
        #if check==1:
        #    print 'подключаюсь'
    
        if check==0:
            html_ads=requests.get(url_ads).text
            soup=BeautifulSoup(html_ads, 'lxml')
            #print i

#area            
           
            try:
                ads_area=soup.find(text='Площадь дома: ').find_parents()[1].text.strip()
                area=ads_area.find(":")+1
                ads_area=float(ads_area[area:-4].strip())
                #print 'площадь-', ads_area
                check=0
            except:
                #print "error" 
                #check=1
                ads_area=0
# floors     
            try:             
                ads_floors=soup.find(text='Этажей в доме: ').find_parents()[1].text.strip()
                floors=ads_floors.find(":")+1
                ads_floors=ads_floors[floors:].strip()
                #ads_floors=codecs.encode(ads_floors, "Windows-1251")
                #print 'этажей-', ads_floors
                check=0
            except:
                #print "error" 
                #heck=1
                ads_floors=0
#land                    
            try:   
                ads_land=soup.find(text='Площадь участка: ').find_parents()[1].text.strip()
                land=ads_land.find(":")+1
                ads_land=ads_land[land:-4].strip()
                #print 'участок-', ads_land
                check=0
            except:
                #print "error" 
                #check=1
                ads_land=0
#centr 
            try:
                ads_centr=soup.find(text='Расстояние до города: ').find_parents()[1].text.strip()
                centr=ads_centr.find(":")+1
                ads_centr=ads_centr[centr:-2].strip()
                #print 'до города-', ads_centr
                check=0
            except:
                #print "error" 
                ads_centr=''
                #check=1
#address
            try:
                ads_address=soup.find_all('div', class_='seller-info-value')[1].text.strip() 
                ads_address=codecs.encode(ads_address, "Windows-1251")
                #print 'адрес-', ads_address 
                check=0
            except:
                #print "error" 
                #check=1
                ads_address=''
#selname        
            try:
                ads_selname=soup.find_all('div', class_='seller-info-value')[0].text.strip()
                selname=ads_selname.find("\n")
                ads_selname=ads_selname[:selname].strip()  
                ads_selname=codecs.encode(ads_selname, "Windows-1251")
                #print 'имя продавца-', ads_selname 
                check=0
            except:
                #print "error" 
                #check=1
                ads_selname=''
#price   
            try:
                ads_price=soup.find_all('div', class_='price-value')[0].text.strip()            
                ads_price=ads_price[:-2].strip() 
                ads_price=int(ads_price.replace(' ',''))
                #ads_floors=codecs.encode(ads_floors, "Windows-1251")
                ads_price_m2=ads_price/ads_area
                #print 'цена-', ads_price
                #print 'цена за м2-',ads_price_m2 
                check=0
            except:
                ads_price=0
                ads_price_m2=0
                #print "error" 
                #check=1
#lat            
            try:
                ads_lat=soup.find('div', class_='b-search-map expanded item-map-wrapper js-item-map-wrapper')['data-map-lat']                   
                #print 'широта-', ads_lat
                check=0
            except:
                #print "error" 
                #check=1
                ads_lat=0
#lon           
            try:
                ads_lon=soup.find('div', class_='b-search-map expanded item-map-wrapper js-item-map-wrapper')['data-map-lon']               
                #print 'долгота-', ads_lon 
                check=0
            except:
                #print "error" 
                #check=1
                ads_lon=0
#descr          
            try:
                ads_descr=[]
                descr=soup.find_all('div', class_='item-description')#.find_all('p')
                for a in descr:
                    ads_descr.append(a.text.strip())
                ads_descr=codecs.encode(ads_descr[0], "Windows-1251")
                #print 'описание-', ads_descr[0]
                check=0
            except:
                #print "error" 
                #check=1
                ads_descr=''
#phone          
            try:
                driver=webdriver.Firefox()
                driver.get(url_ads)

                button = driver.find_element_by_xpath('//button[@class="button item-phone-button js-item-phone-button button-origin button-origin-blue button-origin_full-width button-origin_large-extra item-phone-button_hide-phone item-phone-button_card js-item-phone-button_card"]')
                
                if soup.find(class_='write-message')==None:                    
                    if soup.find(class_='item-tooltip-content').text.strip()!='':                
                        sleep(uniform(3,6))
                        button.click()
                        sleep(uniform(3,6))
                        button.click()                    
                    else:   
                        sleep(uniform(3,6))
                        button.click()  
                else:   
                    sleep(uniform(3,6))
                    button.click()   
                   
                sleep(uniform(3,6))
                driver.save_screenshot('avito_screenshot.png')

                im=driver.find_element_by_xpath('//div[@class="item-phone-big-number js-item-phone-big-number"]//*')
                location=im.location # dict {'x':1234, 'y':1234}
                size=im.size         # dict {'width':1234, 'height':1234}            
                
                driver.quit()
            
                image=Image.open('avito_screenshot.png')
                x=location['x']
                y=location['y']
                width=size['width']
                height=size['height']       
       
                image.crop((x,y, x+width, y+height)).save('tel.gif')

                image=Image.open('tel.gif')
                phone=image_to_string(image)
                phone=phone.replace('-','')
                phone=phone.replace('—','')
                phone=phone.replace(' ','')            
                #print 'телефон-', phone
                    
                check=0                 
            except:
                #print "error" 
                driver.quit()
                phone=0
                #check=1   
            #url_ads=codecs.encode(str(url_ads), "Windows-1251")
            new_line={'price':ads_price,
                      'price_m2':ads_price_m2,
                      'selname':ads_selname,
                      'phone':phone,
                      'address':ads_address,
                      'area':ads_area,
                      'floors':ads_floors,
                      'land':ads_land,
                      'centr':ads_centr,
                      'lat':ads_lat,
                      'lon':ads_lon,
                      'descr':ads_descr,
                      'url':url_ads}
                      
            write_csv(new_line)
    print "обработано объявлений:", i+1, " из:", len(urls)
```
